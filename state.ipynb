```python
import warnings
warnings.filterwarnings("ignore")

import pandas as pd
import numpy as np
import joblib

from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
df = pd.read_csv("Indias_Electricity_Consumption_Dataset.csv")
print(df.head())
```

```python
print(df.columns)
print(df.shape)
```

```python
df = df.drop(columns=[
    "Unnamed: 0",
    "Pondy",
    "Tripura",
    "DD",
    "DNH",
    "Essar steel",
    "DVC"
], errors="ignore")
```

```python
long_df = df.melt(
    id_vars=["Dates"],
    var_name="State",
    value_name="Consumption"
)
```

```python
long_df = long_df[long_df["State"] != "Total Consumption"].copy()
```

```python
long_df["Dates"] = pd.to_datetime(long_df["Dates"])
```

```python
long_df = long_df.sort_values(["State", "Dates"]).reset_index(drop=True)
long_df.head()
```

```python
long_df["year"] = long_df["Dates"].dt.year
long_df["month"] = long_df["Dates"].dt.month
long_df["day"] = long_df["Dates"].dt.day
long_df["dayofweek"] = long_df["Dates"].dt.dayofweek
```

```python
long_df["lag_1"] = long_df.groupby("State")["Consumption"].shift(1)
long_df["lag_2"] = long_df.groupby("State")["Consumption"].shift(2)
long_df["lag_3"] = long_df.groupby("State")["Consumption"].shift(3)
long_df["lag_7"] = long_df.groupby("State")["Consumption"].shift(7)

long_df["rolling_mean_3"] = (
    long_df.groupby("State")["Consumption"]
    .transform(lambda x: x.shift(1).rolling(window=3).mean())
)

long_df["rolling_mean_7"] = (
    long_df.groupby("State")["Consumption"]
    .transform(lambda x: x.shift(1).rolling(window=7).mean())
)
```

```python
long_df = long_df.dropna().reset_index(drop=True)
print(long_df.shape)
```

```python
model_df = pd.get_dummies(long_df, columns=["State"], drop_first=True)
model_df.head()
```

```python
state_cols = [col for col in model_df.columns if col.startswith("State_")]

feature_cols = [
    "year",
    "month",
    "day",
    "dayofweek",
    "lag_1",
    "lag_2",
    "lag_3",
    "lag_7",
    "rolling_mean_3",
    "rolling_mean_7",
] + state_cols

X = model_df[feature_cols]
y = model_df["Consumption"]
```

```python
split_date = pd.Timestamp("2021-07-01")

train_mask = model_df["Dates"] < split_date
test_mask = model_df["Dates"] >= split_date

X_train = X[train_mask]
X_test = X[test_mask]
y_train = y[train_mask]
y_test = y[test_mask]

print("Train shape:", X_train.shape)
print("Test shape:", X_test.shape)
print("Train max date:", model_df.loc[train_mask, "Dates"].max())
print("Test min date:", model_df.loc[test_mask, "Dates"].min())
```

```python
model = RandomForestRegressor(
    n_estimators=150,
    max_depth=30,
    min_samples_split=2,
    min_samples_leaf=1,
    max_features="sqrt",
    random_state=30,
    n_jobs=-1
)

model.fit(X_train, y_train)
```

```python
long_df["lag_14"] = long_df.groupby("State")["Consumption"].shift(14)
```

```python
y_pred = model.predict(X_test)
```

```python
mae = mean_absolute_error(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
r2 = r2_score(y_test, y_pred)

print("Model Performance:")
print(f"MAE  : {mae:.2f}")
print(f"RMSE : {rmse:.2f}")
print(f"R2   : {r2:.4f}")
print(f"Accuracy (R2 %) : {r2*100:.2f}%")
```

```python
joblib.dump(model, "india_state_electricity_model.joblib")
joblib.dump(feature_cols, "india_state_model_columns.joblib")

print("Saved:")
print("- india_state_electricity_model.joblib")
print("- india_state_model_columns.joblib")
```
